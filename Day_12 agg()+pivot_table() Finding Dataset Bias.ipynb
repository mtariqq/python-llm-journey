```python
import pandas as pd
```

```python
ft_data = pd.DataFrame({
    'Prompt_type' : ['code', 'chat', 'code', 'chat', 'code', 'math'],
    'language': ['python', 'english', 'python', 'urdu', 'js', 'english'],
    'quality_score': [4.5, 3.0, 4.8, 2.5, 4.2, 5.0],
    'token_length': [120, 80, 150, 60, 110, 200]
})
```

```python
# Q1: mean, count, min of quality_score per prompt type
q1 = ft_data.groupby('Prompt_type').agg({
    'quality_score': ['mean', 'count', 'min']
})
print("Q1Bias Check\n", q1)
```

```python
# Q2: pivot table of avg quality_score by Pompt_type and language
```

```python
q2 = pd.pivot_table(ft_data,
                    values='quality_score',
                    index='Prompt_type',
                    columns='language',
                    aggfunc='mean')
print("Q2 Pivot check\n", q2)
```
