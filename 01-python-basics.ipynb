```python
# A list of dicts = how finetuning datasets look
dataset = [
    {"instruction": "What is Python?", "output": "A programming language."},
    {"instruction": "What is 2+2?", "output": "4"}
]

# Loop through each example
for example in dataset:
    print("Q:", example["instruction"])
    print("A:", example["output"])
    print("---")
    
```

```python
dataset = [
    {"instruction": "Who is Moeez?", "output": "Moeez is Tariq's son"},
    {"instruction": "what is 3+2?", "output": "5"}
]

for example in dataset:
    print("Q:", example["instruction"])
    print("A:", example["output"])
    print("---")
```

```python
def ask_question(instruction, output):
    print("Q:", instruction)
    print("A:", output)
    print("---")

for example in dataset:
    ask_question(example["instruction"], example["output"])
```
