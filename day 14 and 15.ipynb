```python
import json

train_data = [
{"prompt": "Write a JavaScript for loop from 1 to 5", "completion": "for(let i=1; i<=5; i++) { console.log(i); }"},
{"prompt": "Create a Python list comprehension for squares 1-5", "completion": "squares = [x**2 for x in range(1,6)]"},
{"prompt": "SQL query to select all users over 18", "completion": "SELECT * FROM users WHERE age > 18;"},
{"prompt": "React useState hook for a counter", "completion": "const [count, setCount] = useState(0);"},
{"prompt": "Bash command to list all .txt files", "completion": "ls *.txt"},
{"prompt": "CSS to center a div", "completion": "div { display: flex; justify-content: center; align-items: center; }"}
]

test_data = [
{"prompt": "Write a Python for loop from 1 to 3", "completion": "for i in range(1,4): print(i)"},
{"prompt": "SQL query to get users named John", "completion": "SELECT * FROM users WHERE name = 'John';"},
{"prompt": "JavaScript arrow function that adds two numbers", "completion": "const add = (a, b) => a + b;"}
]
```

```python
with open('train.jsonl', 'w') as f:
    for item in train_data:
        f.write(json.dumps(item) + "\n")

with open('test.jsonl', 'w') as f:
    for item in test_data:
        f.write(json.dumps(item) + "\n")

print("Day 14 files recreated")
!ls train.jsonl test.jsonl

```
