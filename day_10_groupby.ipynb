```python
import pandas as pd
```

```python
data={
    'Student': ['Ali', 'Sara', 'Ahmad', 'Zara', 'Usman', 'Ayesha'],
    'Class': ['A', 'B', 'A', 'B', 'A', 'B'],
     'Marks':[85, 92, 78, 88, 95, 76]
}
df = pd.DataFrame(data)
```

```python
df.groupby('Class')['Marks'].mean()
```

```python
df.groupby('Class')['Student'].count()
```

```python
df.groupby('Class')['Marks'].max()
```

```python
df.groupby('Class')['Marks'].agg(['mean', 'max', 'count'])
```
