```python
def print_trip_plan(plan):
    print("=== Your Trip Plan ===")
    for key in plan:
        print(f"{key}: {plan[key]}")
    print("Planned by:",
plan.get("planner", "Guest"))

trip_plan = {
    "destination": "Hangzhou",
    "days": 7,
    "purpose": "Interview prep",
    "season": "Spring",
    "planner": "Muhammad"
}

print_trip_plan(trip_plan)
```

```python
def print_trip_plan(plan):
    print("=== Your China Trip Plan ===")
    print(f"Main Base: {plan['destination']}")
    print(f"Total Days: {plan['days']}")

    print("\nCities to visit:") # this is our new list
    cities = plan.get("cities_to_visit", []) # get with empty list as default
    for i, city in enumerate(cities, start=1): #enumerate gives us 1,2,3...
        print(f"{i}, {city}")
    print(f"\nPurpose: {plan.get('purpose', 'Travel')}")
    print("Planned by:", plan.get("planner", "Guest"))

# New dict with a LIST inside
trip_plan = {
    "destination": "Hangzhou",
    "days": 3,
    "purpose": "interview perp + Travel",
    "cities_to_visit": ["West Lake", "Alibaba HQ", "Tea Village", "Qiantang River", "Zhejiang Unkiversity"], 
    # <-- List
    "planner": "Muhammad"
}

print_trip_plan(trip_plan)
    
```

```python
print("\n--- Trip Check ---")

if trip_plan["days"] >= 4:
    print("You have time for all cities")
    for i, city in enumerate(trip_plan["cities_to_visit"], start=1):
        print(f"{i}. {city}")

elif trip_plan["days"] >=2:
    print("short trip! Pick your top 2 cities:")
    for i, city in enumerate(trip_plan["cities_to_visit"][:2], start=1): # [:2] = first 2 only
        print(f"{i}. {city}")

else: 
    print("Too Short, for Hangzhou. Pick 1 city or postpone")
```
