```python
students = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Tariq', 'Sadiq', 'Rafiq', 'Sadaf']
})

marks = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks': [88, 92, 86, 99]
})

full_data = pd.merge(students, marks, on='ID')
full_data

```

```python
# Add Asif who has no marks
students = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Tariq', 'Sadiq', 'Rafiq', 'Sadaf', 'Asif']
})

marks = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks' : [88, 92, 86, 99]
})

# Default merge = inner
pd.merge(students, marks, on='ID')
    
```

```python
pd.merge(students, marks, on='ID', how='left')
```

```python
pd.merge(students, marks, on='ID', how='inner')
```

```python
pd.merge(students, marks, on='ID', how='right')
```

```python
pd.merge(students, marks, on='ID', how='outer')
```
